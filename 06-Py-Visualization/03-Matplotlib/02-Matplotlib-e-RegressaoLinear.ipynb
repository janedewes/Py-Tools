## Matplotlib e Regressão Linear

```python
from scipy import stats
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Add documentação no JNB
from IPython.display import Image
Image(url = 'lreg.png')
```

```python
# Preços de casas
x = np.array([113, 347, 199, 371, 549, 301, 419, 579])
# Array numpy
```

```python
# Tamanho de Casas
y = np.array([1119, 1524, 2101, 2232, 2599, 3201, 3687, 4459])
```

```python
# Alg de ML
# O resultado da fç stats.linregress(x, y), gravamos em uma série de variáveis (à esqueda do =)
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
# intercept = a e b
```

```python
plt.plot(x, y, 'ro', color = 'red')
```

```python
plt.ylabel('Tamanho da Casa')
plt.xlabel('Preço')
```

```python
plt.axis([0, 600, 0, 5000])
```

```python
# plt.plot(x, y)
plt.plot(x, x * slope + intercept, 'b')
```

```python
plt.plot()
plt.show()
```
